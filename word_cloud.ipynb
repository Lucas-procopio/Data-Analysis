```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
#example

frase = 'No que diz respeito ao empenho, ao compromisso, ao esforço, à dedicação, não existe meio termo.'
frase = frase + ' Ou você faz uma coisa bem feita ou não faz.'
```

```python
frase
```

```python
# Gerar nuvem de palavras
nuvem_palavras = WordCloud(
    background_color='black',
    width=1600,
    height=600
).generate(frase)
```

```python
nuvem_palavras
```

```python
# Plotting graph

#Defining figure and axis
fig, ax =  plt.subplots( figsize=(10,7))

#calling graph
ax.imshow(nuvem_palavras, interpolation='bilinear')

#Parameter for don't show us axis' graph
ax.set_axis_off()
```
