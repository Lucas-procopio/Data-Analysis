```python
# plotting numpy graphs
np.random.seed(6)
x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

# Defining figure
plt.figure(figsize = (15, 5))

# Defining figure
plt.suptitle('Figura', fontsize = 15)

# plt.subplot(line, total number of columns, number's column)
plt.subplot(1, 3, 1)
# Defining labels
plt.plot(x, y1, color = 'k')
# Include attribute labels
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 2)
plt.plot(x, y2, color = 'r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3)
plt.plot(x, y3, color = 'g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# plt' function to upper space between subplots
plt.tight_layout(pad = 4)

plt.show()
```

```python
# Creating plots, defining 1 line and 3 columns, defining size figure
# Next, assign result to figure
fig, ax = plt.subplots(1, 3, figsize = (15, 5))

# figure's titel 
fig.suptitle('Figure')

# line/axis.adda plot, defining info to axis
ax[0].plot(x, y1, color = 'black')
ax[1].plot(x, y2, color = 'red')
ax[2].plot(x, y3, color = 'green')

# add labels to plot and their titles
for i in range(3):
  ax[i].set(title = f'Subplot {i+1}', xlabel = 'Eixo X', ylabel = 'Eixo Y')
```

```python
# library to import datasets
import plotly.express as px
```

```python
# importing gapminder dataset
px.data.gapminder()
```

```python
# Creating dataframe with a query
df = px.data.gapminder().query('country == "Brazil"').set_index('year')
```

```python
df.head()
```

```python
# Plot data, include index(year) on x axis and using per capita PIB 
plotando os dados, passando o index(ano) no eixo X e passando os dados do PIB per capita no eixo y
plt.plot(df.index, df['gdpPercap'])

# Insert a title on graph
plt.title('PIB per capita do Brasil')

# Title's labels
plt.ylabel('PIB per capita')
plt.xlabel('Tempo');
```

```python
df.head()
```

```python
# Creating title's figure
title = 'Relação entre expectativa de vida e renda per capita no Brasil'

# Creating figure, saying size's figure
plt.figure(figsize = (12, 4))

# Plotting scatter graphs and insert X and Y data
plt.scatter(df['lifeExp'], df['gdpPercap'], cmap = 'viridis')

# Label's titles
plt.xlabel('Expectativa de vida')
plt.ylabel('Renda per capita')

# Title's localization
plt.title(title, loc = 'left')

# Showwing graph
plt.show()
```

```python
# Plotting bar's graph, with X info, color and destribution of population 
plt.bar(x = df.index,
        height = df['pop'],
        color = 'red'
        )

# Insert title's graph
plt.title('População brasileira');
```

```python
# Defining continet filter function
def filtrar_continente(continente):
  # Import dataset
  df = px.data.gapminder()
  # Filter column on dataset
  df = df[df['continent'] == continente]
  return df
```

```python
# Defining country filter function
def filtrar_pais(pais, variavel):
  'Filtra algum pais da amostra para as variáveis "pop", "gdpPercap" e "lifeExp"'
  df = px.data.gapminder()
  # Filter column on dataset
  df = df[df['country'] == pais][variavel] # essa viarável é o segundo parâmetro no filtro, podendo ser outro atributo do dataset
  return df
```

```python
# Filtering continent using continent filter function
americas = filtrar_continente('Americas')
# Filtering country on americanas function and using
# unique method
paises = americas['country'].unique()
# Creating figure (image)
plt.figure(figsize = (12, 8))

# for each country, will be a scatter graph using for function
for pais in paises:
  # filter to country, using filtrar_pais function. For each data of american continent, I will filter one country and one variable
  # (Once to lifeExp variable, another to gdpPercap variable)
  plt.scatter(filtrar_pais(pais= pais,
                           variavel = 'lifeExp'
                          ),
              filtrar_pais(pais = pais,
                           variavel = 'gdpPercap'
                           ))
# Perfumery

# Legend
plt.legend(labels = paises, loc = 'best')
# Geral Title
plt.title('Retenção entre renda per capita e expectativa de vida', loc = 'left')
# axis x title
plt.xlabel('Expectativa de vida')
# axis y title
plt.ylabel('Renda per capita');
```
